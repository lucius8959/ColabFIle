```
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib.font_manager as fm
# path = '/usr/share/fonts/truetype/SimHei.ttf'
# fontprop = fm.FontProperties(fname=zhfont, size=13)

figure = plt.figure(figsize=(8,4),dpi=80)
ax = figure.add_axes((0.1, 0.1, 0.8, 0.8))
X_test1 = [1,5,7,9,13,16]
y_test1 = [15,50,80,40,70,50]

X_test2 = [2,6,8,11,14,16]
y_test2 = [10,40,30,50,80,60]

plt.bar(X_test1,y_test1, color=(1, 0, 0, 1), linewidth=2.0, label='實際值')
plt.bar(X_test2,y_test2, color=(0, 0, 1, 1), linewidth=2.0, label='預測值')

plt.title("模型结果",fontproperties=zhfont)
plt.xlabel("X值",fontproperties=zhfont)
plt.ylabel("Y值",fontproperties=zhfont)
plt.legend(loc = "upper left",prop=zhfont)
plt.show()
```

```
import matplotlib.pyplot as plt

labels = ["東部","南部","北部","西部"]
sizes = [5,10,20,15]
colors = ["red", "green", "blue", "yellow"]
explode = (0,0,0.5,0)

plt.pie(
    sizes, 
    explode = explode, 
    labels=labels, 
    colors = colors, 
    labeldistance=1.1, 
    autopct = "%3.1f%%", 
    shadow=False, 
    startangle=90, 
    center=(10,0), 
    pctdistance=0.6
    )
plt.axis("equal")
plt.legend(loc="upper left",prop=zhfont)
plt.show()
```
